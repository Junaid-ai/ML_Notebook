```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
%matplotlib inline
import numpy as np
#print("All libraries working ✅")
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header = 1)
```

```python
df.info()
```

```python
##data cleaning
#Missing values
#df.isnull().sum()
df[df.isnull().any(axis=1)]
```

```python
df.head()
```

```python
#THE dataset is converted into two sets based region from 122th index, we can make a new 
#column based on the region
# 1: Bejaia Region dataset
#2: sidi-Bel Region dataset
#Add new colum with region

df.loc[:122, "Region"] = 0
df.loc[122:, "Region"] = 1
df.head()
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
#Removing null values
df = df.dropna().reset_index(drop = True)
#df.isnull().sum() #check if null is rmeoved
```

```python
df = df.drop(122).reset_index(drop = True)
```

```python
df.columns
```

```python
#Fix spaces in column name
df.columns = df.columns.str.strip()
```

```python
##change the required columns as integer type data type
#df.columns = df.columns.str.strip()
#df[['day', 'month', 'year', 'Temperature', 'RH','Ws']].astype(int)

df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = \
    df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
#changing other columns to float data type
Objects_str = [features for features in df.columns if df[features].dtypes == 'str']

for i in Objects_str:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
#let save the cleaned datset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv')
```
